```python
import re

file_region_map = {}

for file in file_names:
    # Skip system files like .DS_Store
    if not file.endswith(('.csv', '.json')):
        continue
    
    # Match 2-letter region code at the start of filename (with or without underscore)
    match = re.match(r'^([A-Z]{2})', file)
    
    if match:
        region = match.group(1)
        file_region_map[file] = region
    else:
        file_region_map[file] = "unknown"
```
